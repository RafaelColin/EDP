```python
import networkx as nx  #llamamos a la libreria
```

A continuacion le decimos que vamos hacer un grafo dirigido y le ingresamos la direccion de cada arista

Un grafo dirigido (o dígrafo) es un tipo de grafo en el que cada arista tiene una dirección definida, representada por una flecha que va desde un vértice/nodo origen a un vértice/nodo destino.

```python
G=nx.DiGraph() # Le decimos que es un grafo dirigido
G.add_nodes_from(["A","B", "C", "D"]) #Ingresamos el nombre de cada nodo, se anotan tantos nombres como nodos existan, incluso si no tienen comunicacion con otro nodo.
G.add_edge("A", "B",capacity=35) #Le decimos que la arista parte del nodo A y llega al nodo B.
G.add_edge("A", "C",capacity=35) #Le decimos que la arista parte del nodo A y llega al nodo C.
G.add_edge("B", "D",capacity=35) #Le decimos que la arista parte del nodo B y llega al nodo D.
G.add_edge("C", "D",capacity=35) #Le decimos que la arista parte del nodo C y llega al nodo D.
G.add_edge("B", "C",capacity=35) #Le decimos que la arista parte del nodo B y llega al nodo C.
nx.draw_networkx(G) #Dibuja el grafo con las especificaciones dadas anteriror mente.
```

```python
nx.nodes(G) #Este comando me da una lista con el nombre de cada uno de los nodos
```

Si antes del comando le agrego la sigueinte notacion "print("Nombre de lo que se imprimir", nx.nodes(G))", obtengo un resultado mas estetico como se ve a continuacíon.

```python
print("Nodos",nx.nodes(G))
```

La notacion anterior, brinda un resultado más estetico tambien para los sigueintes comandos, por lo que los escribire de esa forma.

```python
nx.edges(G) #Este comando me da una lista con los arcos ó aristas que estan en las especificaciones iniciales y observables en el grafo.
print("Arcos ó Arsitas",nx.edges(G))
```

```python
nx.number_of_nodes(G) # Este comando me dice la cantidad de nodos
print("Cantidad de nodos",nx.number_of_nodes(G))
```

A continuacion repetire lo mismo que se hizo anteriormente pero en una unica celda, agregando nodos sin arcos y cambiando el nombre de los nodos en opciones numericas y alfa numerica.
# Numerica

```python
G=nx.DiGraph() # Le decimos que es un grafo dirigido
G.add_nodes_from(["1","2", "3", "4", "5"]) #Ingresamos el nombre de cada nodo, se anotan tantos nombres como nodos existan, incluso si no tienen comunicacion con otro nodo.
G.add_edge("1", "2",capacity=35) #La arista parte del nodo 1 y llega al nodo 2.
G.add_edge("1", "3",capacity=35) #La arista parte del nodo 1 y llega al nodo 3.
G.add_edge("2", "3",capacity=35) #La arista parte del nodo 2 y llega al nodo 3.
G.add_edge("2", "4",capacity=35) #La arista parte del nodo 2 y llega al nodo 4.
G.add_edge("3", "4",capacity=35) #La arista parte del nodo 3 y llega al nodo 4.
nx.draw_networkx(G) #Dibuja el grafo con las especificaciones dadas anteriror mente.
print("Nodos",nx.nodes(G)) #Este comando me da una lista con el nombre de cada uno de los nodos y me permite darle un nombre a la lista.
print("Arcos ó Arsitas",nx.edges(G)) #Este comando me da una lista con los arcos ó aristas que estan en las especificaciones iniciales y observables en el grafo;me permite darle un nombre a la lista
print("Cantidad de nodos",nx.number_of_nodes(G))# Este comando me dice la cantidad de nodos y me permite darle un nombre a la lista.
```

# Alfanumerica

```python
G=nx.DiGraph() # Le decimos que es un grafo dirigido
G.add_nodes_from(["1/A","2/B", "3/C", "4/D", "5/E"]) #Ingresamos el nombre de cada nodo, se anotan tantos nombres como nodos existan, incluso si no tienen comunicacion con otro nodo.
G.add_edge("1/A", "2/B",capacity=35) #La arista parte del nodo 1 y llega al nodo 2.
G.add_edge("1/A", "3/C",capacity=35) #La arista parte del nodo 1 y llega al nodo 3.
G.add_edge("2/B", "3/C",capacity=35) #La arista parte del nodo 2 y llega al nodo 3.
G.add_edge("2/B", "4/D",capacity=35) #La arista parte del nodo 2 y llega al nodo 4.
G.add_edge("3/C", "4/D",capacity=35) #La arista parte del nodo 3 y llega al nodo 4.
nx.draw_networkx(G) #Dibuja el grafo con las especificaciones dadas anteriror mente.
print("Nodos",nx.nodes(G)) #Este comando me da una lista con el nombre de cada uno de los nodos y me permite darle un nombre a la lista.
print("Arcos ó Arsitas",nx.edges(G)) #Este comando me da una lista con los arcos ó aristas que estan en las especificaciones iniciales y observables en el grafo;me permite darle un nombre a la lista
print("Cantidad de nodos",nx.number_of_nodes(G))# Este comando me dice la cantidad de nodos y me permite darle un nombre a la lista.
```
