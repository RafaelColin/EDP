#<font color=purple> *Codigo de grafica y=x²*

```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos un array de 400 puntos entre -10 y 10
x = np.linspace(-10, 10, 400)

# Calculamos los valores de y para cada x (y = x²)
y = x**2

# Creamos la figura y los ejes
plt.figure(figsize=(8, 6))

# Dibujamos la gráfica de y = x²
plt.plot(x, y, label='y = x²', color='purple')

# Añadimos título y etiquetas a los ejes
plt.title('Gráfica de y = x²')
plt.xlabel('x')
plt.ylabel('y')

# Agregamos líneas horizontales y verticales en el origen
plt.axhline(0, color='red',linewidth=0.5, ls='--')
plt.axvline(0, color='red',linewidth=0.5, ls='--')

# Agregamos una cuadrícula
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

# Añadimos la leyenda
plt.legend()

# Mostramos la gráfica
plt.show()
```
