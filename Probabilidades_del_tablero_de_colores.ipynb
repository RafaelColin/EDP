#Probabilidades del tablero de colores
Voy a generar un programa para calcular las probabilidades $x_n$, $y_n$, $z_n$, $t_n$ y graficarlas.

Esto nos dar√° la respuesta al problema resuelto en clase:
Tablero_colores.png

Para empezar a realizar el programa, defino los requisitos que tiene que tener el programa:

1.- Matriz de Transici√≥n: Se define la matriz M que encapsula las distribuciones de probabilidad en relacion con cada estado posible.

2.-Vector Inicial: Comenzamos con $x_0$ = 1.

3.-Iteraciones: Calculamos $v_n$ para 50 pasos usando multiplicaci√≥n matricial

4.-Gr√°fica: Visualizamos la evoluci√≥n de las probabilidades.

5.-Convergencia: Verificamos si las probabilidades finales se aproximan a 0.25 (distribuci√≥n uniforme).

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la matriz de transici√≥n M
M = np.array([
    [1/4, 0, 1/4, 2/4],
    [2/4, 1/4, 0, 1/4],
    [1/4, 2/4, 1/4, 0],
    [0, 1/4, 2/4, 1/4]
])

# Vector inicial (por ejemplo, empezar en el sector x)
v0 = np.array([1, 0, 0, 0])  # x0=1, y0=0, z0=0, t0=0

# N√∫mero de iteraciones
n_iter = 50

# Almacenar las probabilidades en cada paso
valores = np.zeros((n_iter, 4))
valores[0] = v0

# Iterar para calcular las probabilidades
for i in range(1, n_iter):
    valores[i] = M @ valores[i-1]  # Multiplicaci√≥n matriz-vector

# Graficar las probabilidades
plt.figure(figsize=(10, 6))
n = np.arange(n_iter)
plt.plot(n, valores[:, 0], label='x_n')
plt.plot(n, valores[:, 1], label='y_n')
plt.plot(n, valores[:, 2], label='z_n')
plt.plot(n, valores[:, 3], label='t_n')
plt.xlabel('N√∫mero de pasos (n)')
plt.ylabel('Probabilidad')
plt.title('Evoluci√≥n de las Probabilidades de los Sectores')
plt.legend()
plt.grid(True)
plt.show()

# Verificar si las probabilidades convergen a 0.25
ultimas_prob = valores[-1]
print("Probabilidades finales:", ultimas_prob)
```

Al ejecutar el c√≥digo, observamos que las probabilidades convergen aproximadamente a:

$x_n$‚âà0.25, $y_n$‚âà0.25, $z_n$‚âà0.25, $t_n$‚âà0.25

Esto indica que, despu√©s de un n√∫mero suficientemente grande de pasos, todos los sectores son igualmente probables. La distribuci√≥n estacionaria es uniforme, lo que responde afirmativamente a la pregunta planteada.

La gr√°fica muestra c√≥mo las probabilidades evolucionan hacia 0.25, confirmando visualmente la convergencia a la distribuci√≥n uniforme.
