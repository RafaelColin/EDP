```python
import networkx as nx  #llamamos a la libreria
```

A continuacion le decimos que vamos hacer un grafo dirigido y le ingresamos la direccion de cada arista

Un grafo dirigido (o dígrafo) es un tipo de grafo en el que cada arista tiene una dirección definida, representada por una flecha que va desde un vértice/nodo origen a un vértice/nodo destino.

```python
G=nx.DiGraph() # Le decimos que es un grafo dirigido
G.add_nodes_from(["A","B", "C", "D"]) #Ingresamos el nombre de cada nodo, se anotan tantos nombres como nodos existan, incluso si no tienen comunicacion con otro nodo.
G.add_edge("A", "B",capacity=25) #Le decimos que la arista parte del nodo A y llega al nodo B.
G.add_edge("A", "C",capacity=5) #Le decimos que la arista parte del nodo A y llega al nodo C.
G.add_edge("B", "D",capacity=35) #Le decimos que la arista parte del nodo B y llega al nodo D.
G.add_edge("C", "D",capacity=15) #Le decimos que la arista parte del nodo C y llega al nodo D.
G.add_edge("B", "C",capacity=3) #Le decimos que la arista parte del nodo B y llega al nodo C.
nx.draw_networkx(G) #Dibuja el grafo con las especificaciones dadas anteriror mente.
```

Los comandos que pongo a continuacion nos permiten saber la informacion basica del grafo:

-Nombre de los nodos.

-Arcos o aristas existentes.

-Cantidad de nodos.

-Cantidad de arcos o aristas.

```python
nx.nodes(G) #Este comando me da una lista con el nombre de cada uno de los nodos
```

La notacion anterior, brinda un resultado más estetico tambien para los sigueintes comandos, por lo que los escribire de esa forma.

```python
nx.edges(G) #Este comando me da una lista con los arcos ó aristas que estan en las especificaciones iniciales y observables en el grafo.
print("Arcos ó Arsitas",nx.edges(G)) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

```python
nx.number_of_nodes(G) # Este comando me dice la cantidad de nodos
print("Cantidad de nodos",nx.number_of_nodes(G)) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

```python
nx.number_of_edges(G) # Este comando me dice la cantidad de arcos ó aristas
print("Cantidad de arcos ó aristas",nx.number_of_edges(G)) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

El grado de un nodo es el número de conexiones o aristas que tiene con otros nodos en una red o grafo. En redes no dirigidas, es simplemente el total de aristas conectadas a él, mientras que en redes dirigidas se distingue entre el grado de entrada (aristas que apuntan al nodo) y el grado de salida (aristas que parten del nodo).

<font color = red> Grado en redes no dirigidas:

Es la cantidad de aristas que están conectadas a un nodo específico.

<font color = red>Grado en redes dirigidas:

-Grado de entrada (In-degree): El número de aristas que apuntan o llegan hacia el nodo.

-Grado de salida (Out-degree): El número de aristas que parten o salen del nodo.

-Grado total: La suma del grado de entrada y el grado de salida.

A continuacion veremos como obtener el grado total de cada nodo.

```python
nx.degree(G) #Este comando me da una lista con el grado total de cada uno de los nodos
print("Grado",nx.degree(G))
```

El "camino más corto" es un problema que busca la ruta de menor "peso" (que puede ser distancia, tiempo o costo) entre dos o más puntos, a continuacion les dare un comando con el que podemos abtener el dato extrallendolo desde el grafo.

```python
shortest_path = nx.shortest_path(G,source="A",target="D") #Este comando busca el camino mas corto entre dos nodos dados.
print("Camino mas corto",shortest_path) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

La centralidad de nodos en el análisis de redes mide la importancia de un nodo dentro de una red, utilizando diferentes métricas como la centralidad de grado, que cuenta las conexiones; la centralidad de cercanía, que evalúa la distancia promedio a otros nodos; la centralidad de intermediación, que mide cuántas rutas cortas pasa por el nodo; y la centralidad de vector propio, que considera la influencia de los nodos conectados.
A continuacion unos comando que nos dan las medidas basicas centralidad.

```python
centralidad_grado = nx.degree_centrality(G) #Este comando da la centralidad de grado, cuantas conecxiones directas tiene un nodo.
print("Centralidad",centralidad_grado) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

```python
centralidad_cercania = nx.closeness_centrality(G) #Este comando da la centralidad de cercania, la distancia promedio a otros nodos.
print("Centralidad de grado",centralidad_cercania) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

```python
centralidad_intermediacion = nx.betweenness_centrality(G) #Este comando da la centralidad de intermediacion, cuantas rutas cortas pasa por el nodo.
print("Centralidad de intermediacion",centralidad_intermediacion) #Imprime la respuesta del comando de la linea superior y me permite darle un nombre a la lista.
```

```python
centralidad_vector_propio = nx.eigenvector_centrality(G, max_iter=1000) #Este comando da la centralidad de vector propio, considera la influencia de los nodos conectados.
#Es similar a la centralidad por grado, pero le da mas importancia a estar conectado con nodos que tambien son importantes.
print("Centralidad de vector propio",centralidad_vector_propio) #Imprime la respuesta del comando de la linea y me permite darle un nombre a la lista.
```

Un árbol de expansión mínima es un subconjunto de aristas de un grafo no dirigido y conectado que conecta todos los nodos de forma que la suma de los pesos de las aristas sea la menor posible, sin formar ciclos.

```python
# Árbol de expansión mínima
print("\n" + "="*50)
print("ÁRBOL DE EXPANSIÓN MÍNIMA:")
G_no_dirigido = G.to_undirected()
for u, v in G_no_dirigido.edges():
    if G.has_edge(u, v) and 'capacity' in G[u][v]:
        G_no_dirigido[u][v]['weight'] = G[u][v]['capacity']
mst = nx.minimum_spanning_tree(G_no_dirigido)
print("Peso total del MST:", mst.size(weight='weight'))
print("Aristas en el MST:", list(mst.edges()))
```

La ruta más corta es el camino de menor costo o distancia entre dos puntos en una red, y se determina mediante algoritmos como el de Dijkstra, que es fundamental en aplicaciones de transporte, telecomunicaciones y planificación.

```python
# Ruta más corta con pesos
print("\n" + "="*50)
print("RUTA MÁS CORTA CON PESOS:")
try:
    shortest_path_weighted = nx.shortest_path(G, source="A", target="D", weight='capacity')
    path_length = nx.shortest_path_length(G, source="A", target="D", weight='capacity')
    print("Ruta más corta considerando capacidades:", shortest_path_weighted)
    print("Longitud de la ruta:", path_length)
except:
    print("No se pudo calcular la ruta con pesos")
```

El modelo de flujo máximo es una técnica de optimización utilizada para determinar la mayor cantidad de flujo que puede pasar a través de una red desde un origen hasta un destino, utilizando nodos, arcos y capacidades.

```python
# Flujo máximo
print("\n" + "="*50)
print("FLUJO MÁXIMO:")
flow_value, flow_dict = nx.maximum_flow(G, "A", "D")
print("Flujo máximo de A a D:", flow_value)
print("Flujo en cada arco:")
for u in flow_dict:
    for v, flow in flow_dict[u].items():
        if flow > 0:
            capacity = G[u][v]['capacity']
            print(f"  {u} → {v}: {flow}/{capacity}")
```
